# Text Classification with TFâ€‘IDF + Logistic Regression

Toy sentiment analysis with a few short text samples.

```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

texts = [
    'I love this product',
    'Really hate the experience',
    'Amazing quality and great support',
    'Terrible, will not buy again',
    'Very good, I am happy',
    'Worst purchase ever'
]
labels = [1,0,1,0,1,0]

df = pd.DataFrame({'text': texts, 'label': labels})
print(df)
```

```
tfidf = TfidfVectorizer()
X = tfidf.fit_transform(df['text'])
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

model = LogisticRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)
acc = accuracy_score(y_test, pred)
print('Accuracy:', acc)

for text, true_label, pred_label in zip(df['text'].iloc[y_test.index], y_test, pred):
    print(f'Text: {text!r} | True: {true_label} | Pred: {pred_label}')
```
